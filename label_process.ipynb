```python
import pandas as pd
data_label=pd.read_table("train_labels.txt",sep="\t")
```

```python
type_mapping = {    
    'Business':0,
    'Sci/Tech':1,
    'World':2,
    'Sports':3
}
```

```python
data_label.columns=['label']
```

```python
data_label['label']=data_label['label'].map(type_mapping)
```

```python
data_label.to_csv("train_labels2.csv")
```

```python
data_label
```

```python
data_text=pd.read_table("train_texts.txt",sep="\t")
data_text.columns=['text']
```

```python
data_text
```

```python
data_train=pd.concat([data_text,data_label],axis=1,ignore_index=True)
```

```python
data_train.columns=['text','label']
```

```python
data_train
```

```python
data_train.to_csv("data_train.txt")
```
